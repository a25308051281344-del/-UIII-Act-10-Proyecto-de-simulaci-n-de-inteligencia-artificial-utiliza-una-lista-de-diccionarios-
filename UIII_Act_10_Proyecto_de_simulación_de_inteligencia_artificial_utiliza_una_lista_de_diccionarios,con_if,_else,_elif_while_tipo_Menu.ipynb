```python
print("======================================")
print("   IA DE DISEÑADORES VIRTUALES")
print("======================================")

# Lista vacía
lista_disenadores = []

# Ciclo principal
while True:

    print("\n========= MENÚ =========")
    print("1. Agregar diseñador")
    print("2. Mostrar diseñadores")
    print("3. Buscar diseñador")
    print("4. Actualizar diseñador")
    print("5. Eliminar diseñador")
    print("6. Simulación IA")
    print("7. Salir")

    opcion = input("Selecciona una opción: ")

    # AGREGAR
    if opcion == "1":

        # Diccionario vacío
        disenador = {}

        nombre = input("Nombre del diseñador: ")
        especialidad = input("Especialidad: ")
        experiencia = input("Años de experiencia: ")
        estilo = input("Estilo favorito: ")

        disenador["nombre"] = nombre
        disenador["especialidad"] = especialidad
        disenador["experiencia"] = experiencia
        disenador["estilo"] = estilo

        lista_disenadores.append(disenador)

        print("\nDiseñador agregado correctamente.")

    # MOSTRAR
    elif opcion == "2":

        if len(lista_disenadores) == 0:
            print("\nNo hay diseñadores registrados.")
        else:
            print("\n===== LISTA DE DISEÑADORES =====")

            contador = 1

            for d in lista_disenadores:

                print("\nDiseñador", contador)
                print("Nombre:", d["nombre"])
                print("Especialidad:", d["especialidad"])
                print("Experiencia:", d["experiencia"])
                print("Estilo:", d["estilo"])

                contador += 1

    # BUSCAR
    elif opcion == "3":

        buscar = input("Ingresa el nombre a buscar: ")

        encontrado = False

        for d in lista_disenadores:

            if d["nombre"] == buscar:

                print("\nDiseñador encontrado")
                print("Nombre:", d["nombre"])
                print("Especialidad:", d["especialidad"])
                print("Experiencia:", d["experiencia"])
                print("Estilo:", d["estilo"])

                encontrado = True

        if encontrado == False:
            print("\nDiseñador no encontrado.")

    # ACTUALIZAR
    elif opcion == "4":

        actualizar = input("Nombre del diseñador a actualizar: ")

        encontrado = False

        for d in lista_disenadores:

            if d["nombre"] == actualizar:

                print("\nEscribe los nuevos datos")

                d["nombre"] = input("Nuevo nombre: ")
                d["especialidad"] = input("Nueva especialidad: ")
                d["experiencia"] = input("Nueva experiencia: ")
                d["estilo"] = input("Nuevo estilo: ")

                print("\nDatos actualizados correctamente.")

                encontrado = True

        if encontrado == False:
            print("\nDiseñador no encontrado.")

    # ELIMINAR
    elif opcion == "5":

        eliminar = input("Nombre del diseñador a eliminar: ")

        encontrado = False

        for d in lista_disenadores:

            if d["nombre"] == eliminar:

                lista_disenadores.remove(d)

                print("\nDiseñador eliminado correctamente.")

                encontrado = True

        if encontrado == False:
            print("\nDiseñador no encontrado.")

    # SIMULACIÓN IA
    elif opcion == "6":

        print("\n===== IA RECOMENDADORA =====")

        estilo_usuario = input("¿Qué estilo te gusta? ")

        print("\nLa IA recomienda estos diseñadores:\n")

        encontrado = False

        for d in lista_disenadores:

            if d["estilo"] == estilo_usuario:

                print("Nombre:", d["nombre"])
                print("Especialidad:", d["especialidad"])
                print("---------------------------")

                encontrado = True

        if encontrado == False:
            print("La IA no encontró diseñadores con ese estilo.")

    # SALIR
    elif opcion == "7":

        print("\nGracias por usar la IA de Diseñadores Virtuales.")
        break

    # OPCIÓN INCORRECTA
    else:
        print("\nOpción inválida.")
```
